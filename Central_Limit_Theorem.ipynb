# Introduction

The central limit theorem tells us that the sampling distribution of the sample mean 
is normally distributed with a mean equal to the mean of the underlying population.

We will want to see if our computed sample means indeed cluster around the pouplation mean.

We will sample 35 values from a uniform distribution on the interval \[-10,10\],
and then compute the sample mean for each sample; we will perform this procedure 20 times, and then plot a histogram of our sample means.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#In this cell we compute the 20 different sample means.

means = []

for i in range(20):
    sample = np.random.uniform(-10,10,35)
    sample_mean = sum(sample)/len(sample)
    
    means.append(sample_mean)
    
    
```

```python
#In this cell we plot our 20 sample means and the population mean

fig, ax = plt.subplots()


plt.xlim(-11,11)

plt.vlines(0,ymin = 0,ymax = 5,color = 'r')
plt.hist(means)


plt.show()
```

We see that the sample means are clustered around the population mean 0, even though the underlying data comes from a uniform distribution.

An example of a data set sampled from a uniform distribution likes the following:

```python
#In this cell we show an example of 35 data points sampled from a uniform distribution on the interval [-10,10]

dataset = np.random.uniform(-10,10,35)

fig,ax = plt.subplots()

plt.xlim(-11,11)

plt.vlines(0,ymin = 0,ymax = 5,color = 'r')

ax.hist(dataset)

plt.show()
```

We see that the underlying distribution of the data and the sample mean hsitograms are different.

# Conclusion

In conclusion, we see that the sample means computed from uniformily distributed data cluster around the population mean.
